# Genetic algorithm

We will try to implement genetic algorithm for producing string, e. g. 'Hello world'

```python
import numpy as np
import matplotlib.pyplot as plt
import string
import random
from pprint import pprint
import math
string.ascii_letters = string.ascii_letters + ' '
```

```python
class StringFitnessFunction:
    def __init__(self, string: str = 'Hello world'):
        self.string = string
        
    
    def __call__(self, x):
        error_count = 0
        for i, (ch1, ch2) in enumerate(zip(self.string, x)):
            if ch1 != ch2:
                error_count += 1
        error_count += max(len(self.string), len(x)) - i - 1
        return error_count
```

```python
string_to_converge = 'My name is Ilya'
max_length = len(string_to_converge) + 2 # arbitrary choosen
fitness = StringFitnessFunction(string_to_converge)
assert fitness(string_to_converge) == 0
assert max_length > len(string_to_converge)
```

```python
def generate_population(size):
    population = []
    for _ in range(size):
        length = np.random.randint(1, max_length)
        entity = ''
        for i in range(length):
            entity = entity + random.choice(string.ascii_letters)
        population.append(entity)
    return population
```

```python
population = generate_population(100)
pprint(population[0:10])
```

```python
fitnesses = [fitness(x) for x in population]
idx_best = np.argmin(fitnesses)
print('Best score is {}: "{}".'.format(fitnesses[idx_best], population[idx_best]))
```

```python
def get_best_n(population, n):
    fitnesses = np.array([fitness(x) for x in population])
    best_n_idxs = np.argsort(fitnesses)
    best_n = np.array(population)[best_n_idxs[:n]]
    return best_n.tolist()

assert len(get_best_n(population, 10)) == 10
```

```python
def reproduce(population, reprod_num):
    def cross(parent_1, parent_2):
        common_length = min(len(parent_1), len(parent_2))
        appendix = parent_1[common_length:] + parent_2[common_length:]
        appendix = appendix[:math.floor(len(appendix) / 2)]
        cross_point = random.randint(0, common_length - 1)
        child_1 = parent_1[:cross_point] + parent_2[cross_point:]
        child_2 = parent_2[:cross_point] + parent_1[cross_point:] + appendix
        return child_1, child_2
    childs = []
    for _ in range(reprod_num):
        parent_1, parent_2 = random.choice(population), random.choice(population)
        child_1, child_2 = cross(parent_1, parent_2)
        childs.extend([child_1, child_2])
    population.extend(childs)
    random.shuffle(population)
```

```python
before = set(population)
reproduce(population, 10)
after = set(population)
print('Number of childs: {}'.format(len(after - before)))
```

```python
def mutation(population, percent):
    for i in range(len(population)):
        if np.random.rand() > percent:
            continue
        mutated = population[i]
        for j in range(len(population[i])):
            if np.random.rand() < 0.5:
                mutated = mutated[:j] + random.choice(string.ascii_letters) + mutated[j+1:]
        population[i] = mutated
        
        # add / reduce letters
        extra = np.random.randint(-2, 2)
        if extra < 0:
            population[i] = population[i][:-extra]
        else:
            for _ in range(extra):
                population[i] = population[i] + random.choice(string.ascii_letters)
```

```python
mutation_dynamic = []
percent = 0.1
for _ in range(100):
    before = set(population)
    mutation(population, percent)
    after = set(population)
    mutation_percent = len(after - before) / len(before)
    mutation_dynamic.append(mutation_percent)

print('Mutation percent in average: {}'.format(sum(mutation_dynamic) / len(mutation_dynamic)))
plt.plot(mutation_dynamic, color='black')
plt.fill_between(np.arange(0, len(mutation_dynamic)), 0, mutation_dynamic, color = 'green')
plt.xlabel('time')
plt.ylabel('mutation percent')
plt.title('Mutation dynamic')
plt.show()
```
